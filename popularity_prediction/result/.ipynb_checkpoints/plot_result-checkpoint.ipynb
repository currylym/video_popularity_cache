```python
# 读取结果
import pandas as pd

def init():
    key1 = ['historyBased','mlr','svr','mse-True-0','mse-False-0','mse-True-1','mse-False-1',
                                 'mse-True-2','mse-False-2','mse-True-3','mse-False-3']
    key2 = ['mse','mae']
    res = {}
    for i in key1:
        res[i] = {j:[] for j in key2}
    return res

def read_result(path,res):
    with open(path) as f:
        for line in f:
            line = line.strip().split(' ')
            line = [i for i in line if i]
            method,mse,mae = line
            if method != 'method':
                res[method]['mse'].append(mse)
                res[method]['mae'].append(mae)
res = init()
read_result('HistoryNum=7_UpdateCycle=6H/2016-10-3_2016-10-23_2016-10-24_2016-10-30_default_1.txt',res)
print(res)
```

```python
# 对默认参数下的结果进行分析
import os
import numpy as np
result_dir = 'HistoryNum=7_UpdateCycle=6H/'

def read_result_dir(path):
    result = init()
    for file in os.listdir(path):
        if file.split('.')[-1] == 'txt':
            read_result(path+file,result)

    def mean(nums):
        nums = list(map(float,nums))
        return np.mean(nums)

    def std(nums):
        nums = list(map(float,nums))
        return np.std(nums)

    for method in result:
        result[method]['mse-mean'] = mean(result[method]['mse'])
        result[method]['mae-mean'] = mean(result[method]['mae'])
        result[method]['mse-std'] = std(result[method]['mse'])
        result[method]['mae-std'] = std(result[method]['mae'])
    
    result = pd.DataFrame(result)
    new_col = {'mse-True-0':'True-0',
               'mse-False-0':'False-0',
               'mse-True-1':'True-1',
               'mse-False-1':'False-1',
               'mse-True-2':'True-2',
               'mse-False-2':'False-2',
               'mse-True-3':'True-3',
               'mse-False-3':'False-3'}
    result = result.rename(columns=new_col)
    return result.T

result = read_result_dir(result_dir)
result.head()
```

```python
import matplotlib.pyplot as plt
def box_plot(df):
    plt.figure()
    plt.title('mse')
    mse = []
    for method in df.index:
        nums = df.loc[method,'mse']
        nums = list(map(float,nums))
        mse.append(nums)
    plt.boxplot(mse[3:],labels=list(df.index)[3:],vert=True)
    plt.xticks(rotation='60')
    for method in ['mlr']:
        value = df.loc[method,'mse-mean']
        plt.hlines(value,1,8,label=method,linestyle='--',color='r')
    plt.grid(axis='y')
    plt.legend()
    plt.ylabel('mse')
    
    plt.figure()
    plt.title('mae')
    mae = []
    for method in df.index:
        nums = df.loc[method,'mae']
        nums = list(map(float,nums))
        mae.append(nums)
    plt.boxplot(mae[3:],labels=list(df.index)[3:],vert=True)
    plt.xticks(rotation='60')
    for method in ['mlr']:
        value = df.loc[method,'mae-mean']
        plt.hlines(value,1,8,label=method,linestyle='--',color='r')
    plt.grid(axis='y')
    plt.legend()
    plt.ylabel('mae')
box_plot(result)
```

```python
# 对比attention模式在不同采样周期下的性能提升比例
def improvment_percent(true,false):
    return (float(false)-float(true))/float(false)

def improvment_plot(results,labels,xlabel,evaluation):
    
    plt.figure()
    All = [] 
    for result in results:
        true_result = []
        false_result = []
        for method in result.index:
            if 'True' in method:
                true_result.extend(result.loc[method,evaluation])
            if 'False' in method:
                false_result.extend(result.loc[method,evaluation])
        improvment = [improvment_percent(i,j) for i,j in zip(true_result,false_result)]
        All.append(improvment)
    #print(true_result)
    #print(false_result)
    plt.boxplot(All,labels=labels)
    #plt.xticks(rotation='60')
    plt.xlabel(xlabel)
    plt.ylabel('improvment percent')
    plt.grid(axis='y')
    plt.title(evaluation)
    
improvment_plot([result1,result,result2],['3H','6H','12H'],'update cycle','mse')
improvment_plot([result1,result,result2],['3H','6H','12H'],'update cycle','mae')
improvment_plot([result3,result,result4],['5','7','9'],'observation cycles','mse')
improvment_plot([result3,result,result4],['5','7','9'],'observation cycles','mae')
```
